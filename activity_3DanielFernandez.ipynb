The Trans-Atlantic Slave Trade Database represents an exceptional digital repository that provides one of the most comprehensive and meticulously organized collections of information regarding the transatlantic slave trade. Drawing upon decades of scholarly research, it unites the contributions of researchers, archivists, and technology experts to reconstruct the extensive scale of forced human migration experienced by African captives destined for the Americas. This database not only chronicles the journeys of enslaved individuals—including the ships, routes, and traders involved—but also emphasizes the significant moral, cultural, and geopolitical changes that transformed the world from the 16th to the 19th centuries. By the year 1820, the number of enslaved Africans crossing the Atlantic had greatly surpassed that of European migrants, establishing the transatlantic slave trade as the largest and most consequential forced migration in human history up to that point. This resource provides critical insights into the ways in which the slave trade was interwoven with European colonial expansion, the decline of indigenous populations, and the economic development of the Americas. Created through collaborations among esteemed institutions such as Emory University, the University of California, and Harvard’s Hutchins Center, and supported by the National Endowment for the Humanities, this initiative has assembled experts from various disciplines to develop a tool for historical reflection.

While the Trans-Atlantic Slave Trade Database does not encompass a complete account of the millions of enslaved Africans transported across the Atlantic, it serves as an essential entry point by documenting the experiences of over 91,000 individuals whose personal information was preserved in 19th-century naval records. Although these records do not fully represent the identities of all enslaved people, they offer valuable data for comprehending the vastness, logistics, and human toll of the trade. With more than 36,000 documented voyages, the database enables scholars to investigate the connections between the slave trade, warfare, political turmoil, and environmental factors.

### 4. What is the overall average proportion of ```percent_women```, ```percent_children```, ```percent_men```?

````python
voyages_df[['percent_women', 'percent_children', 'percent_men']].mean()
````

### 5. How many of the column values for ```percent_women```, ```percent_children```, ```percent_men``` are left blank? Suggest 1 reason why majority of the values in these columns are blank.

````python
voyages_df[['percent_women', 'percent_children', 'percent_men']].isna().value_counts()
````

When isolating the values, we see that 17,828 columns do not contain a recorded value meaning that they were left blank. The reason, as noted in the text, is that most journeys failed to document details about the individuals they were transporting. I would also speculate that this lack of records stems from the fact that enslavers saw the people they abducted as commodities rather than human beings. As a result, gathering personal information likely did not occur to those who focused on recording the economic aspects of the voyages.
